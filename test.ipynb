```python
import scipy.optimize as opt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('./coral_Useries.csv')
data
```

```python
d234Um = np.asarray(data['d234Um'])
d234Ul = np.asarray(data['d234Ui'])
U238 = np.asarray(data['238U(ppm)'])
Th232 = np.asarray(data['232Th(ppt)'])
Th230_U238_ratio = np.asarray(data['230Th/238U'])
```

```python
def rhs(t, d234Um):
    lam230 = 1/245000
    lam234 = 1/75000
    dlam = lam230 - lam234
    return -np.exp(-lam230*t) + d234Um/1000*(lam230/dlam)*(1-np.exp(-dlam*t))
```

```python
for i in range(np.shape(Th230_U238_ratio)[0]):
    t = opt.fsolve(rhs, Th230_U238_ratio[i]-1, args=(d234Um[i]))
    print(t)
```
